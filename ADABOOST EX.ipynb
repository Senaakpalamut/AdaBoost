```python
#ADABOOST
#1. IMPORTING THE MODULES
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split #to split the data into training and testing
from sklearn import metrics
```

```python
#2. DATA EXPLORATION
iris = datasets.load_iris()
x = iris.data #features such as sepal and petal (lenght,width)
y= iris.target #types such as virginia, setosa, versicolor
print(x)
print(y)
```

```python
#3. SPLITTING THE DATA
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.4)
#splitting the datasets as 60% training and 40% testing.
```

```python
#4. FITTING THE MODEL
function= AdaBoostClassifier(n_estimators = 1000, learning_rate=20)
model=function.fit(x_train,y_train)
```

```python
#5. MAKING PREDICTIONS ABOUT OUR TRAGETED VALUES
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#6. MODEL EVALUATION
print('Accuracy:', metrics.accuracy_score(y_test,y_pred)*100,'%')
#it is to tell how many times our model predicts the correct classes.
```
